```python
# Definición de la clase
class Entrenador:
    # El método __init__ es el constructor que inicializa los atributos
    def __init__(self, nombre, deporte, años_experiencia):
        self.nombre = nombre                # Atributo 1
        self.deporte = deporte              # Atributo 2
        self.años_experiencia = años_experiencia  # Atributo 3

    # Función 1: Calentar
    def calentar(self):
        print(f"{self.nombre} está dirigiendo el calentamiento de {self.deporte}. ¡A mover los músculos!")

    # Función 2: Explicar ejercicio
    def explicar_ejercicio(self):
        print(f"{self.nombre} está explicando la rutina técnica. Pongan atención a los detalles.")

    # Función 3: Motivar
    def motivar(self):
        print(f"Dice {self.nombre}: '¡Vamos equipo! Sus {self.años_experiencia} años de experiencia le dicen que pueden lograrlo.'")

# --- Creación de Objetos (Instancias) ---

# Creando el entrenador 1 con datos de ejemplo
entrenador1 = Entrenador("Carlos Ruiz", "Fútbol", 10)

# Creando el entrenador 2 con datos de ejemplo
entrenador2 = Entrenador("Elena Spark", "Natación", 5)

# --- Uso de los Objetos ---

print("--- Actividades del Entrenador 1 ---")
entrenador1.calentar()
entrenador1.explicar_ejercicio()
entrenador1.motivar()

print("\n--- Actividades del Entrenador 2 ---")
entrenador2.calentar()
entrenador2.explicar_ejercicio()
entrenador2.motivar()
print("programa terminado")
print("creador de software por Ramirez Azul NL = 44")
```
