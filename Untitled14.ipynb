#### Clase Juguete
####Clase Juguete con 3 atributos (nombre, edadRecomendada, material), 2 funciones leerDatos(), mostrarDatos() y 2 objetos (juguete1, juguete2).
####que sea interactivo --> los input
####crear 2 objetos juguete1, juguete2
####no usar while true, if, for, listas y diccionarios

![Imagen Juguete](https://raw.githubusercontent.com/RamirezAlvarezazul128/ExaUII-Clase-Juguete-Ramirez-Azul-0441/refs/heads/main/infoclasejuguete.png)

```python

class Juguete:
    def __init__(self, nombre, edadRecomendada, material):
        self.nombre = nombre
        self.edadRecomendada = edadRecomendada
        self.material = material

    def leerdatos(self):
        self.nombre = input("Ingrese el nombre del juguete: ")
        self.edadRecomendada = input("Ingrese la edad recomendada: ")
        self.material = input("Ingrese el material del juguete: ")

    def mostrardatos(self):
        print(f"Nombre: {self.nombre}")
        print(f"Edad Recomendada: {self.edadRecomendada}")
        print(f"Material: {self.material}")


print("Juguete Numero 1")
juguete1 = Juguete("", "", "") # Creando el objeto juguete1
juguete1.leerdatos()
juguete1.mostrardatos()
print("--*-*-*-*-*-*-*-*-*")


print("Juguete Numero 2")
juguete2 = Juguete("", "", "") # Creando el objeto juguete2
juguete2.leerdatos()
juguete2.mostrardatos()
print("--*-*-*-*-*-*-*-*-*")

print("software desarrollado por Ramirez Azul NL = 44")
```
